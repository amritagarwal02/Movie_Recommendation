```python
df.dropna(inplace=True)
```

```python
movies=df[["id","title","overview","genre","original_language"]]
movies
```

```python
movies["tags"]=movies["original_language"]+" "+movies["genre"]+" "+movies["overview"]
```

```python
movies
```

```python
movies_df=movies.drop(["genre","overview","original_language"],axis=1)
movies_df
```

```python
movies_df.iloc[3715]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector=cv.fit_transform(movies_df["tags"].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar=cosine_similarity(vector)
```

```python
similar[0][0:3]
```

```python
distance=sorted(list(enumerate(similar[1])),reverse=True,key=lambda vector:vector[1])
```

```python
for i in distance[0:6]:
    print(movies_df.iloc[i[0]].title)
```

```python
def recommend(movies):
    index=movies_df[movies_df["title"]==movies].index[0]
    distance=sorted(list(enumerate(similar[index])),reverse=True,key=lambda vector:vector[1])
    for i in distance[0:6]:
        print(movies_df.iloc[i[0]].title)
```

```python
recommend("Dilwale Dulhania Le Jayenge")
```
